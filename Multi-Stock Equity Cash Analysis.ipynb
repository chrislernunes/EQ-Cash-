#### 1. Data collecting from yahoo finance Api

```python
#Top 10 blue chip stock list
stock_list = ["HDFCBANK.NS", "ITC.NS", "LT.NS", "HDFCAMC.NS", "AXISBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS", "SBIN.NS", "TATASTEEL.NS", "TCS.NS"]
#stock history
stock_insight = []
#creating for loop to iterate in stock list
for i in stock_list:
        #data mining from yahoo api
        data = yf.Ticker(i.upper())
        #setting 60d chart
        stock = data.history(period="365d")
        #storing history of stock
        stock_insight.append(stock)
```

```python
'''Once the ticker data is stored into variable that it can be pass through this algoriths for 7days insight'''

def seven_days(i):
    #Checking minimum
    open_minimum = i.tail(7).describe()[3:4]['Open']
    close_minimum = i.tail(7).describe()[3:4]['Close']
    print(f'7days chart Open Minimum: {open_minimum[0]}')
    print()
    print(f'7days chart Closed Minimum: {close_minimum[0]}')
    
    #Checking high
    open_maximum = i.tail(7).describe()[7:8]['Open']
    close_maximum = i.tail(7).describe()[7:8]['Close']
    print(f'7days chart Open Minimum: {open_maximum[0]}')
    print()
    print(f'7days chart Closed Minimum: {close_maximum[0]}')
    
    #Checking high
    open_maximum = i.tail(7).describe()[7:8]['Open']
    close_maximum = i.tail(7).describe()[7:8]['Close']
    print(f'7days chart Open Minimum: {open_maximum[0]}')
    print()
    print(f'7days chart Closed Minimum: {close_maximum[0]}')
    print()
    print(f'Stock mean Open price was {open_mean[0]} Rupees in 7days chart. Minimum Open price was {open_minimum[0]} Rupees, and Minimum Open price was {open_maximum[0]}. similarly, Closing mean price was {close_mean[0]} Rupees in 7days chart, Minimum Closed price was {close_minimum[0]} Rupees and Maximum Close price was {close_maximum[0]} Rupees')
```

#### Algorithms will take stock ticker as input and plot all graph.

```python
def technical_analysis(i):
    sns.set_style('whitegrid')
    plt.figure(figsize=(10,5))
    sns.lineplot(data=i, x="Date", y="Open", color="black")
    plt.ylabel("Open price ()")
    plt.xlabel("Date")
    plt.show()
    
    plt.figure(figsize=(10,5))
    x = i["Open"].index
    fit = np.polyfit(x, i["Open"], deg=1)
    fit_function = np.poly1d(fit)
    sns.scatterplot(data=i["Open"], color="black")
    sns.lineplot(x=x, y=fit_function(x), color="red")
    plt.xlabel("Year")
    plt.ylabel("Price ()")
    plt.show()
    
    features = ['Open', 'High', 'Low', 'Close', 'Volume']
 
    plt.subplots(figsize=(20,10))
 
    for j, col in enumerate(features):
        plt.subplot(2,3,j+1)
        sns.distplot(i[col])
    plt.show()
    
    def volume_analaysis(i):
            open, close, volume, date = i["Open"], i["Close"], i["Volume"], i.index.values
            result = open, close, volume, date 
            sns.set_style('whitegrid')
            plt.figure(figsize=(8,5))
            plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
            plt.scatter(date, volume)
            plt.xlabel("Date")
            plt.ylabel("Volume millions")
            plt.show()
```

#### 9. Independent Index and Dependent stock

```python
#defining function
def stock_analysis(stock1, stock2):
    #list of arguments
    stock_list = [stock1, stock2]
    #history chart
    history = []
    #creating for loop for calling tiacker
    for i in stock_list:
            #data mining from yahoo api
            data = yf.Ticker(i.upper())
            #setting 60d chart
            stock = data.history(period="365d")
            #storing history of stock
            history.append(stock)
    #returning output
    return history
```

```python
#calling function
result = stock_analysis("RELIANCE.NS", "^NSEI")
#defining stock1 and stock2
stock1, stock2 = result[0], result[1]
#creating dataset
```

### Algorithms for prediction best fit line, which can take two assets, x is Independent and y is Dependent variable

```python
'''This ML algorithms will compare dependent varibale which is y and Independent varibale x, this norally take ticke and try to predict  best fit line'''
def index(stock1, stock2):
    stock1 = yf.Ticker(stock1)
    stock2 = yf.Ticker(stock2)
    stock1_history = stock1.history(period="365d")
    stock2_history = stock2.history(period="365d")

    reg = np.polyfit(stock2_history["Open"], stock1_history["Open"], deg=1)
    print(reg)
    #prediction x axis is indepent and y axis is dependent
    #array([l, m]) #Every l percent of stock2 will make m percent increase in stock1

    trend = np.polyval(reg, stock2_history["Open"])
    plt.figure(figsize=(20,10))
    plt.scatter(stock2_history["Open"], stock1_history["Open"], color="black")
    plt.plot(stock2_history["Open"], trend, "r")
    plt.ylabel("STOCK1 Price")
    plt.xlabel("STOCK2 Price")
    plt.show()
```

### 9 Volume Analysis

#### 9.1 SBIN volume

```python
def volume_analaysis(stock1):
        #data mining from yahoo api
        data = yf.Ticker(stock1.upper())
        #setting 60d chart
        stock = data.history(period="60d")
        open, close, volume, date = stock["Open"], stock["Close"], stock["Volume"], stock.index.values
        return open, close, volume, date
```

```python
result = volume_analaysis("SBIN.NS")
open, close, volume, date = result[0], result[1], result[2], result[3]
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.scatter(date, volume)
plt.xlabel("Date")
plt.ylabel("Volume millios")
plt.show()
```
