```python
plt.figure(figsize=(15,10))
sns.scatterplot(x="Date", y="Open", data = df_SNAP_five, color="red")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("SNAP")
plt.show()
```

## SNAP TREND LINE

```python
plt.figure(figsize=(15,10))
x = np.arange(df_SNAP_five["Open"].size)
fit = np.polyfit(x, df_SNAP_five["Open"], deg=1)
fit_function = np.poly1d(fit)
sns.lineplot(data=df_SNAP_five["Open"], color="red")
sns.lineplot(x=df_SNAP_five["Open"].index, y=fit_function(x))
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("SNAP")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x="Date", y="Open", data = df_snp500_five, color="blue")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("SNP500")
plt.show()
```

## SNP500 TREND LINE

```python
plt.figure(figsize=(15,10))
x = np.arange(df_snp500_five["Open"].size)
fit = np.polyfit(x, df_snp500_five["Open"], deg=1)
fit_function = np.poly1d(fit)
sns.lineplot(data=df_snp500_five["Open"], color="red")
sns.lineplot(x=df_snp500_five["Open"].index, y=fit_function(x))
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("SNP500")
plt.show()
```

## PRICE MOMENT OF SNAP AND SNP500 IN LAST FIVE YEARS

```python
fig, ax1 = plt.subplots(figsize=(15,10))

# plot y1 on the first axis
ax1.plot(x, y1, color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('SNAP', color='red')

# create a second axis that shares the same x-axis
ax2 = ax1.twinx()

# plot y2 on the second axis
ax2.plot(x, y2, color='blue')
ax2.set_ylabel('SNP500', color='blue')

plt.show()
```

```python
reg = np.polyfit(df_snp500_five["Open"],df_SNAP_five["Open"], deg=1) #prediction x axis is snp500 and y axis is SNAP
reg #array([l, m]) #Every l percent of SNP500 will make m percent increase in SNAP
```

```python
plt.figure(figsize=(15,10))
trend = np.polyval(reg, df_snp500_five["Open"])
plt.scatter(df_snp500_five["Open"],df_SNAP_five["Open"])
plt.plot(df_snp500_five["Open"], trend, "r")
plt.ylabel("SNAP")
plt.xlabel("SNP500")
plt.show()
```
