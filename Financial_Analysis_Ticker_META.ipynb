#### Import Libraries

```python
import yfinance
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import html5lib
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", None)
```

#### DS & Algorithms

```python
stock_ticker = input("Enter Your Stock Ticker: ").upper()
stock = yfinance.Ticker(stock_ticker)
history_stock = stock.history(period="730d")
stock_data = pd.DataFrame(history_stock)
financials = stock.financials
df_financials = pd.DataFrame(financials)
financials_qtr= stock.quarterly_financials
df_financials_qtr = pd.DataFrame(financials_qtr)
balance_sheet= stock.balance_sheet
df_balance_sheet = pd.DataFrame(balance_sheet)
quarterly_balance_sheet = stock.quarterly_balance_sheet
df_quarterly_balance_sheet = pd.DataFrame(quarterly_balance_sheet)
cashflow = stock.cashflow
df_cashflow = pd.DataFrame(cashflow)
quarterly_cashflow = stock.quarterly_cashflow
df_quarterly_cashflow = pd.DataFrame(quarterly_cashflow)
earnings = stock.earnings
df_earnings = pd.DataFrame(earnings)
quarterly_earnings = stock.quarterly_earnings
df_quarterly_earnings = pd.DataFrame(quarterly_earnings)
sustainability = stock.sustainability
df_sustainability = pd.DataFrame(sustainability)
calendars = stock.calendar
df_calendars = pd.DataFrame(calendars)
options = stock.options
df_options = pd.DataFrame(options)
news = stock.news
df_news = pd.DataFrame(news)
institutional_holders = stock.institutional_holders
df_institutional_holders = pd.DataFrame(institutional_holders)
major_holders = stock.major_holders
df_major_holders = pd.DataFrame(major_holders)
splits = stock.splits
df_splits = pd.DataFrame(splits)
dividends = stock.dividends
df_dividends = pd.DataFrame(dividends)
actions= stock.actions
df_actions = pd.DataFrame(actions)
```

#### <font color="blue">Graph 1 -  Annual Research and Development Analysis<font>

```python
stock = yfinance.Ticker(stock_ticker)
stock = stock.financials
rnd_2018 = float(stock.iloc[0:1,3])
rnd_2019 = float(stock.iloc[0:1,2])
rnd_2020 = float(stock.iloc[0:1,1])
rnd_2021 = float(stock.iloc[0:1,0])
rnd = {f"{stock_ticker} 2018":rnd_2018,f"{stock_ticker} 2019":rnd_2019,f"{stock_ticker} 2020":rnd_2020,f"{stock_ticker} 2021":rnd_2021 }
if rnd_2021 > rnd_2020:
    message_1 = ("\nResearch and development has been increased")
    print(message_1)
elif rnd_2021 < rnd_2020:
    message_2 = ("\nResearch and development has been decreased")
    print(message_2)
plotdata = pd.DataFrame({f"{stock_ticker}":[(rnd_2018), (rnd_2019), (rnd_2020), (rnd_2021)]}, index=["2018","2019","2020", "2021"])
plotdata.plot(kind="bar")
plotdata[f"{stock_ticker}"].plot(kind="line", color=["red"], figsize=(10,8))
plt.title(f"{stock_ticker}")
plt.xlabel("Research Development")
plt.ylabel("Dollars ($)")
plt.show()
```

#### Research and development validation

```python
print(f"R&D 2018: ${rnd_2018}")
print(f"R&D 2019: ${rnd_2019}")
print(f"R&D 2020: ${rnd_2020}")
print(f"R&D 2021: ${rnd_2021}")
```

```python
print(f"Difference on Annual R&D: ${rnd_2021 - rnd_2020}")
```

#### <font color="blue">Graph 2 - Annual Revenue<font>

```python
stock = yfinance.Ticker(stock_ticker)
stock = stock.financials
revenue_1 = float(stock.iloc[15:16,3])
revenue_2 = float(stock.iloc[15:16,2])
revenue_3 = float(stock.iloc[15:16,1])
revenue_4 = float(stock.iloc[15:16,0])
rnd = {f"{stock_ticker} 2018":revenue_1,f"{stock_ticker} 2019":revenue_2,f"{stock_ticker} 2020":revenue_3,f"{stock_ticker} 2021":revenue_4 }
if revenue_4 > revenue_3:
    increase_revenue = revenue_4 - revenue_3
    message_1 = (f"\nRevenue has been increases ${increase_revenue}")
    print(message_1)
elif rvenue_4 < revenue_3:
    decrease_revenue= revenue_4 - revenue_3
    message_2 = (f"\nRevenue has been decreased ${decrease_revenue}")
    print(message_2)
plotdata = pd.DataFrame({f"{stock_ticker}":[(revenue_1), (revenue_2), (revenue_3), (revenue_4)]}, index=["2018","2019","2020", "2021"])
plotdata.plot(kind="bar")
plotdata[f"{stock_ticker}"].plot(kind="line", color=["red"], figsize=(10,8))
plt.title(f"{stock_ticker}")
plt.xlabel("Revenue Generated")
plt.ylabel("Dollars ($)")
plt.show()
```
