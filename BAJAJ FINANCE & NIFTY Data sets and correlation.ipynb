## BAJAJ FINANCE TREND LINE

```python
plt.figure(figsize=(15,10))
x = np.arange(df_BAJFINANCE_five["Open"].size)
fit = np.polyfit(x, df_BAJFINANCE_five["Open"], deg=1)
fit_function = np.poly1d(fit)
sns.lineplot(data=df_BAJFINANCE_five["Open"], color="purple")
sns.lineplot(x=df_BAJFINANCE_five["Open"].index, y=fit_function(x))
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("BAJAJ FINANCE")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x="Date", y="Open", data = df_NIFTY50_five, color="black")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("BAJAJ FINANCE")
plt.show()
```
